# Welcome to Danbury AI's June Workshop!
In this workshop we will use neural networks to produce models of four different datasets in order to provide a glimpse into the different ways neural networks can be used. 

* [The Titanic Disaster](https://www.kaggle.com/andrewrib/dai-june-2018-workshop-notebook-1): How to use a *Feed-Forward Neural Network* to predict the survivors of the Titanic based on categorical ( gender, ticket class, etc ) and quantitative ( age, ticket price, etc) features. 
* [MNIST](https://www.kaggle.com/andrewrib/dai-june-2018-workshop-notebook-2): How to use a *Feed-Forward Neural Network* to classify well preprocessed images of hand written digits.
* [Street View House Numbers](https://www.kaggle.com/andrewrib/dai-june-2018-workshop-notebook-4): How to use a *Convolutional Feed-Forward Neural Network* to classify noisy real world images of google street view house numbers. 
* [Jigsaw Toxic Comment Classification Challenge](https://www.kaggle.com/andrewrib/dai-june-2018-workshop-notebook-5): How to use a *Long Short-Term Memory Network* to classify wikipedia comments. 

We will use an in-class competition to organize these notebooks in the kernels section, but the actual competition is only tied to the [Street View House Numbers](http://ufldl.stanford.edu/housenumbers/) dataset. It is hoped that the competition portion of this workshop will give you a chance to experience the process of solving real world machine learning problems and an opportunity to collaborate with other Danbury AI members and the Kaggle community at large.  We will be hosting the competition till our July meeting and will feature the winners solution in our [Danbury AI newsletter](http://eepurl.com/dxufLf).
